## Flight Price Prediction - Data Collection

**Problem Statement**

**Anyone who has booked a flight ticket knows how unexpectedly the prices vary. The cheapest
available ticket on a given flight gets more and less expensive over time. This usually happens as
an attempt to maximize revenue based on -**

**1. Time of purchase patterns (making sure last-minute purchases are expensive)**

**2. Keeping the flight as full as they want it (raising prices on a flight which is filling up in order
to reduce sales and hold back inventory for those expensive last-minute expensive
purchases)**

**So, you have to work on a project where you collect data of flight fares with other features and
work to make a model to predict fares of flights.**

You have to scrape at least 1500 rows of data. You can scrape more data as well. In this section you have to scrape the data of flights from different websites. The number of columns for data doesn’t havelimit, it’s up to you and your creativity. Generally, these columns areairline name, date of journey,source, destination, route, departure time, arrival time, duration, total stops and the target variable price. You can make changes to it, you can add or you can remove some columns, it completely depends on the website from which you are fetching the data.

```python
#importing the libraries

import selenium
from selenium import webdriver

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By

from selenium.common.exceptions import StaleElementReferenceException
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementNotInteractableException
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import WebDriverException
from selenium.common.exceptions import StaleElementReferenceException

import pandas as pd
import re
import time

import warnings
warnings.filterwarnings('ignore')
```

We are collecting the flight ticket price data from www.ixigo.com. We are collecting the flight data and price for popular cities in India:
    
`New Delhi`, `Mumbai`, `Chennai`, `Bangalore`, `Kochi`

Details to be scraped:
    
`Airline Name`   `Date of Journey`  `Source`    `Destination`    `Departure Time`    `Arrival Time`     `Duration`     `Total Stops`    `Price`

#### Collecting Data from Ixigo.com

```python
Locations = ['New Delhi', 'Mumbai', 'Chennai', 'Bangalore', 'Kochi']

airline_name = []
date_of_journey = []
source_place = []
destination = []
departure_time = []
arrival_time = []
duration = []
total_stops = []
flight_price = []

delay = 10

airline_temp = []
date_temp = []
source_temp = []
destination_temp = []
departure_time_temp = []
arrival_time_temp = []
duration_temp = []
total_stops_temp = []
price_temp = []
```

```python
#User defined functiton for scraping the data.

def flight_data_temp(a,b):
    
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument("--incognito")

    driver = webdriver.Chrome("chromedriver.exe", chrome_options = chrome_options)
    driver.maximize_window()
    time.sleep(1)
    
    global airline_temp_1, date_temp_1, source_temp_1, destination_temp_1, departure_time_temp_1, arrival_time_temp_1, duration_temp_1, total_stops_temp_1, price_temp_1
    
    airline_temp_1 = []
    date_temp_1 = []
    source_temp_1 = []
    destination_temp_1 = []
    departure_time_temp_1 = []
    arrival_time_temp_1 = []
    duration_temp_1 = []
    total_stops_temp_1 = []
    price_temp_1 = []
    
    for sour in Locations[a:b]:
        for destin in Locations:
            if sour!=destin:
                driver.close()


                chrome_options = webdriver.ChromeOptions()
                chrome_options.add_argument("--incognito")

                driver = webdriver.Chrome("chromedriver.exe", chrome_options = chrome_options)
                driver.maximize_window()
                time.sleep(1)

                for tr in range(1,6):
                    for td in range(1,8):
                        if (tr ==1 and td==1):
                            pass
                        elif (tr ==5 and td>=3):
                            pass
                        else:
                            try:
                                time.sleep(1)
                                driver.get("https://www.ixigo.com")
                                time.sleep(1)
                            except:
                                time.sleep(3)
                                driver.get("https://www.ixigo.com")
                                time.sleep(1)

                            #Clearning Search
                            try:
                                clear_search = driver.find_element_by_xpath("//div[@class = 'clear-input ixi-icon-cross']")
                                clear_search.click()
                            except (NoSuchElementException, WebDriverException) as e:
                                pass

                            #selecting the source city
                            try:
                                source_city = driver.find_element_by_css_selector("div.form-cntr.flight > div > div > div.orgn.u-ib.u-v-align-bottom.u-text-left > div > div.c-input-cntr > input")
                                source_city.clear()
                                source_city.send_keys(sour)
                            except WebDriverException as e:
                                time.sleep(2)
                                source_city = driver.find_element_by_css_selector("div.form-cntr.flight > div > div > div.orgn.u-ib.u-v-align-bottom.u-text-left > div > div.c-input-cntr > input")
                                source_city.clear()
                                source_city.send_keys(sour)

                            try:
                                WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.CSS_SELECTOR,'div.result-row.flight-airport.u-box-result.selected > div > div.city-row > div')))
                                select_source_city = driver.find_element_by_css_selector("div.result-row.flight-airport.u-box-result.selected > div > div.city-row > div")
                                select_source_city.click()
                            except ElementNotInteractableException as e:
                                time.sleep(1)
                                source_city.click()

                                select_source_city = driver.find_element_by_xpath("//*[@id='content']/div/div[1]/div[6]/div/div/div[1]/div/div[3]/div/div[1]/div/div[1]/div")
                                select_source_city.click()

                            #Selecting the destination city
                            dest_city = driver.find_element_by_css_selector("div.form-cntr.flight > div > div > div.dstn.u-ib.u-v-align-bottom.u-text-left > div > div.c-input-cntr.focus > input")
                            dest_city.send_keys(destin)

                            WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.CSS_SELECTOR,'div.autocompleter-results.u-box.autocompleter-results-caret > div > div.result-row.flight-airport.u-box-result.selected > div > div.city-row > div')))

                            time.sleep(1)
                            try: 
                                select_dest_city = driver.find_element_by_css_selector("div.autocompleter-results.u-box.autocompleter-results-caret > div > div.result-row.flight-airport.u-box-result.selected > div > div.city-row > div")
                                select_dest_city.click()
                            except ElementNotInteractableException as e:
                                dest_city.click()
                                select_dest_city = driver.find_element_by_xpath("//*[@id='content']/div/div[1]/div[6]/div/div/div[3]/div/div[3]/div/div[1]/div/div[1]/div")
                                select_dest_city.click()

                            #Date Selector
                            journey_date_input = driver.find_element_by_css_selector("div.form-cntr.flight > div > div > div.date.u-ib.u-text-left > div > div.from-date.u-ib > div > input")
                            journey_date_input.click()

                            WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.CLASS_NAME,"rd-month-label")))
                            time.sleep(1)
                            try:
                                select_date = driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[2]/table/tbody/tr[1]/td[2]/div[1]")
                                select_date.click()
                            except ElementNotInteractableException as e:
                                select_date = driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[1]/table/tbody/tr[{}]/td[{}]/div[1]".format(tr,td))
                                select_date.click()

                            search_button = driver.find_element_by_css_selector("div.form-cntr.flight > div > div > div.search.u-ib.u-v-align-bottom > button > div")
                            search_button.click()

                            time.sleep(1)
                            try:
                                WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.CSS_SELECTOR,'div.flight-info.flight-info-no-burn > div.flight-summary > div.left-wing > div.date')))
                            except TimeoutException as e:
                                time.sleep(2)
                            
                            
                            
                            airline_temp = []
                            date_temp = []
                            source_temp = []
                            destination_temp = []
                            departure_time_temp = []
                            arrival_time_temp = []
                            duration_temp = []
                            total_stops_temp = []
                            price_temp = []

                            try:
                                try:
                                    WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.CSS_SELECTOR,'div.flight-info.flight-info-no-burn > div.airline-info > div.text > div > a > div')))
                                    airline_tag = driver.find_elements_by_xpath("//a[@class = 'flight-name']/div")
                                    for air in airline_tag:
                                        airline_temp.append(air.text)
                                except NoSuchElementException as e:
                                    airline_temp.append("-")
                            except StaleElementReferenceException as e:
                                WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.CLASS_NAME,'flight-name')))
                                airline_tag = driver.find_elements_by_xpath("//a[@class = 'flight-name']/div")
                                for air in airline_tag:
                                    airline_temp.append(air.text)

                            try:
                                WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.CSS_SELECTOR,'div.date')))
                                date_tag = driver.find_elements_by_css_selector("div.flight-info.flight-info-no-burn > div.flight-summary > div.left-wing > div.date")
                                for date in date_tag:
                                    date_temp.append(date.text)
                            except NoSuchElementException as e:
                                date_temp.append("-")

                            try:
                                WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.CSS_SELECTOR,'div.city.u-text-ellipsis')))
                                source_tag = driver.find_elements_by_css_selector("div.flight-info.flight-info-no-burn > div.flight-summary > div.left-wing > div.city.u-text-ellipsis")
                                for source in source_tag:
                                    source_temp.append(source.text)
                            except NoSuchElementException as e:
                                source_temp.append("-")

                            try:
                                WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.CSS_SELECTOR,'div.city.u-text-ellipsis')))
                                dest_tag = driver.find_elements_by_css_selector("div.flight-info.flight-info-no-burn > div.flight-summary > div.right-wing > div.city.u-text-ellipsis")
                                for dest in dest_tag:
                                    destination_temp.append(dest.text)
                            except NoSuchElementException as e:
                                destination_temp.append("-")

                            try:
                                WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.CSS_SELECTOR,'div.time')))
                                depart_tag = driver.find_elements_by_css_selector("div.flight-info.flight-info-no-burn > div.flight-summary > div.left-wing > div.time")
                                for depart in depart_tag:
                                    departure_time_temp.append(depart.text)
                            except NoSuchElementException as e:
                                departure_time_temp.append("-")

                            try:
                                WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.CSS_SELECTOR,'div.time')))
                                arrive_tag = driver.find_elements_by_css_selector("div.flight-info.flight-info-no-burn > div.flight-summary > div.right-wing > div.time")
                                for arrive in arrive_tag:
                                    arrival_time_temp.append(arrive.text)
                            except NoSuchElementException as e:
                                arrival_time_temp.append("-")

                            try:
                                WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.CSS_SELECTOR,'div:nth-child(2)')))
                                duration_tag = driver.find_elements_by_css_selector("div.flight-info.flight-info-no-burn > div.flight-summary > div.timeline-widget.u-lfloat > div > div:nth-child(2)")
                                for dur in duration_tag:
                                    duration_temp.append(dur.text)
                            except NoSuchElementException as e:
                                duration_temp.append("-")

                            try:
                                WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.CSS_SELECTOR,'div:nth-child(6)')))
                                stops_tags = driver.find_elements_by_css_selector("div.flight-info.flight-info-no-burn > div.flight-summary > div.timeline-widget.u-lfloat > div > div:nth-child(6)")
                                for stops in stops_tags:
                                    total_stops_temp.append(stops.text)
                            except NoSuchElementException as e:
                                total_stops_temp.append("-")

                            try:
                                WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.CSS_SELECTOR,'span:nth-child(2)')))
                                price_tags = driver.find_elements_by_css_selector("div.provider-list.provider-no-burn > div > div > div > div.price > div > span:nth-child(2)")
                                for price in price_tags:
                                    price_temp.append(price.text)
                            except NoSuchElementException as e:
                                price_temp.append("-")

                            for air,date,sourc,destine,departure,arrival,durat,total,prices in zip(airline_temp,date_temp,source_temp,destination_temp,departure_time_temp,arrival_time_temp,duration_temp,total_stops_temp,price_temp):
                                airline_temp_1.append(air)
                                date_temp_1.append(date)
                                source_temp_1.append(sourc)
                                destination_temp_1.append(destine)
                                departure_time_temp_1.append(departure)
                                arrival_time_temp_1.append(arrival)
                                duration_temp_1.append(durat)
                                total_stops_temp_1.append(total)
                                price_temp_1.append(prices)

                                
    

    print(len(airline_temp_1))
    print(len(date_temp_1))
    print(len(source_temp_1))
    print(len(destination_temp_1))
    print(len(departure_time_temp_1))
    print(len(arrival_time_temp_1))
    print(len(duration_temp_1))
    print(len(total_stops_temp_1))
    print(len(price_temp_1))
```

```python
#Used defined function for saving the scraped data into variables
def flight_data():
    
    for air,date,sourc,destine,departure,arrival,durat,total,prices in zip(airline_temp_1, date_temp_1, source_temp_1, destination_temp_1, departure_time_temp_1, arrival_time_temp_1, duration_temp_1, total_stops_temp_1, price_temp_1):
        airline_name.append(air)
        date_of_journey.append(date)
        source_place.append(sourc)
        destination.append(destine)
        departure_time.append(departure)
        arrival_time.append(arrival)
        duration.append(durat)
        total_stops.append(total)
        flight_price.append(prices)

                                
                                

    print(len(airline_name))
    print(len(date_of_journey))
    print(len(source_place))
    print(len(destination))
    print(len(departure_time))
    print(len(arrival_time))
    print(len(duration))
    print(len(total_stops))
    print(len(flight_price))
```
