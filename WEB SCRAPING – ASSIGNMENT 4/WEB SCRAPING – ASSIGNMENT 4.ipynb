### 1. Scrape the details of most viewed videos on YouTube from Wikipedia.

Details to be scraped:
    
A) `Rank`
B) `Name`
C) `Artist`
D) `Upload date`
E) `Views`

```python
#Importing the libraries

import pandas as pd
import re
import selenium
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
from selenium.common.exceptions import ElementNotInteractableException
from selenium.common.exceptions import NoSuchElementException

from selenium.webdriver.chrome.options import Options     #importing Options
chrome_options = Options()
chrome_options.add_argument('--incognito')

import warnings
warnings.filterwarnings('ignore')
```

```python
chrome_options.add_argument('--incognito')
driver = webdriver.Chrome('chromedriver.exe')

driver.maximize_window() 
time.sleep(1)
```

```python
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
#Extracting the details from the website.

rank = []
name = []
artist = []
date = []
views = []

#Extracting the rank details
rank_tags = driver.find_elements_by_xpath("//table[@class = 'wikitable sortable jquery-tablesorter']/tbody/tr/td[1]")
for ra in rank_tags:
    rank.append(ra.text)
rank = rank[:30]
print(len(rank))

#Extracting the name details
name_tags = driver.find_elements_by_xpath("//table[@class = 'wikitable sortable jquery-tablesorter']/tbody/tr/td[2]")
for na in name_tags:
    name.append(na.text[:-3].replace('[',''))

name = name[:30]
print(len(name))

#Extracting the artist details
artist_tags = driver.find_elements_by_xpath("//table[@class = 'wikitable sortable jquery-tablesorter']/tbody/tr/td[3]")
for art in artist_tags:
    artist.append(art.text)
artist = artist[:30]
print(len(artist))

#Extracting the upload date details
date_tags = driver.find_elements_by_xpath("//table[@class = 'wikitable sortable jquery-tablesorter']/tbody/tr/td[5]")
for dt in date_tags:
    date.append(dt.text)
date = date[:30]
print(len(date))

#Extracting the Views details
view_tags = driver.find_elements_by_xpath("//table[@class = 'wikitable sortable jquery-tablesorter']/tbody/tr/td[4]")
for vi in view_tags:
    views.append(vi.text)
views = views[:30]
print(len(views))

#Creating the dateframe to display the scraped content
youtube_top = pd.DataFrame({})
youtube_top['Rank'] = rank
youtube_top['Video Name'] = name
youtube_top['Artist/Uploader'] = artist
youtube_top['Upload Date'] = date
youtube_top['Views(In Billions)'] = views

youtube_top
```

```python
driver.close()
```

### 2. Scrape the details team India’s international fixtures from bcci.tv.

Details to be scraped

A) `Match title (I.e. 1st ODI)`
B) `Series`
C) `Place`
D) `Date`
E) `Time`

```python
import pandas as pd

import selenium
from selenium import webdriver
import time

import warnings
warnings.filterwarnings('ignore')
```

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.maximize_window()
time.sleep(1)
```

```python
driver.get('http://www.bcci.tv/')
```

```python
#Selecting the category - International
inter = driver.find_element_by_xpath("//li[@class = 'nav-item ml-0']")
inter.click()
```

```python
#Extracting the details

titles = []
series = []
place = []
date = []
time = []

#Extracting match titles
title_tags = driver.find_elements_by_xpath("//span[@class = 'matchOrderText ng-binding ng-scope']")
for ti in title_tags:
    titles.append(ti.text.replace(' -',''))
print(len(titles))

#Extracting series details
series_tags = driver.find_elements_by_xpath("//span[@class = 'ng-binding']")
for se in series_tags:
    series.append(se.text)
print(len(series))

#Extracting place detailis
place_tags = driver.find_elements_by_xpath("//div[@class = 'fix-place ng-binding ng-scope']")
place_tag = driver.find_elements_by_xpath("//div[@class = 'fix-place ng-binding ng-scope']/span[2]")
for pl,pt in zip(place_tags,place_tag):
    place.append(pt.text+' '+pl.text.split()[-1])
print(len(place))

#Extracting date details
date_tags = driver.find_elements_by_xpath("//div[@class = 'match-card-left match-schedule']")
for dt in date_tags:
    date.append(dt.text)
print(len(date))

#Extracting time details
time_tags = driver.find_elements_by_xpath("//div[@class = 'match-card-right match-schedule ']")
for ti in time_tags:
    time.append(ti.text)
print(len(time))

#Creating the dateframe to display the scraped content
ind_int_fix = pd.DataFrame({'Match Title':titles,'Series': series,'Place':place,'Date':date,'Time':time})
ind_int_fix
```

### 5.Scrape the details of trending repositories on Github.com.

Details to be scraped

A) `Repository title`
B) `Repository description`
C) `Contributors count`
D) `Language used`

```python
#importing libraries

import pandas as pd
import re

import selenium
from selenium import webdriver
import time

from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementNotInteractableException

import warnings
warnings.filterwarnings('ignore')
```

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.maximize_window()
time.sleep(1)
```

```python
driver.get('https://github.com/')
time.sleep(2)
```

```python
try:
    #Clicking the category - Explore
    explore_click = driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details")
    explore_click.click()

    #Clicking the category - Trending
    trend_click = driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a")
    trend_click.click()
except ElementNotInteractableException:
    #Clicking the category - Explore
    explore_click = driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details")
    explore_click.click()

    #Clicking the category - Trending
    trend_click = driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a")
    trend_click.click()
```

```python
#Extracting the url for every repositories
url = []
url_tags = driver.find_elements_by_xpath("//h1[@class = 'h3 lh-condensed']/a")
time.sleep(2)
for ur in url_tags:
    url.append(ur.get_attribute('href'))
print(len(url))

titles = []
description = []
contribution = []
language = []

for u in url:
    driver.get(u)
    time.sleep(1)
    
    #Extracting the repository titles
    try:
        titles.append(driver.find_element_by_xpath("//strong[@class = 'mr-2 flex-self-stretch']/a").text)
    except NoSuchElementException:
        titles.append('-')
        
    
    #Extracting Repository Description
    try:
        description.append(driver.find_element_by_xpath("//div[@class = 'BorderGrid-row hide-sm hide-md']/div/p").text)
    except NoSuchElementException:
        description.append('-')
    
    #Extracting Contribution Count
    try:
        contribution.append(driver.find_element_by_xpath("//h2[@class='h4 mb-3']/a[contains(text(),'Contributors')]/span").text)
    except NoSuchElementException:
        contribution.append('-')

    #Extracting Languages used
    lan = []
    try:
        language_tags = driver.find_elements_by_xpath("//li[@class='d-inline']//a//span[1]")
        for la in language_tags:
            lan.append(la.text)
    except NoSuchElementException:
        pass
    language.append(str(lan).replace("'","")[1:-1])  
  
    
print(len(titles))
print(len(description))
print(len(contribution))
print(len(language))



github_trending = pd.DataFrame({'Repository Title':titles,'Repository Description':description,'Contributors Count':contribution,'Languages Used':language})
github_trending
```

```python
driver.close()
```

```python
#Clicking the category - View all Data Sets
all_datasets = driver.find_element_by_xpath("/html/body/table[1]/tbody/tr/td[2]/span[2]/a/font/b")
all_datasets.click()
```

```python
#Scraping the information

name = []
data_type = []
task = []
attribute_type = []
instances = []
attributes = []
year = []

#Scraping the dataset names
name_tags  = driver.find_elements_by_xpath("//tr/td[2]/p/b/a")
for na in name_tags:
    name.append(na.text)
print(len(name))

#Scraping the data type information
dtype_tags = driver.find_elements_by_xpath("//table[2]/tbody/tr/td[2]/p")
for dt in dtype_tags:
    data_type.append(dt.text)
data_type = data_type[1:]
print(len(data_type))

#Scraping the task information
task_tags = driver.find_elements_by_xpath("//tbody/tr/td[3]/p")
for ta in task_tags:
    task.append(ta.text)
task = task[1:-1]
print(len(task))

#Scraping the attribute type information
attribute_type_tags = driver.find_elements_by_xpath("//tbody/tr/td[4]/p")
for at in attribute_type_tags:
    attribute_type.append(at.text)
attribute_type = attribute_type[1:]
print(len(attribute_type))

#Scraping instances information
instance_tags = driver.find_elements_by_xpath("//tbody/tr/td[5]/p")
for inst in instance_tags:
    instances.append(inst.text)
instances = instances[1:]
print(len(instances))

#Scraping attributes information
attribute_tags = driver.find_elements_by_xpath("//tbody/tr/td[6]/p")
for att in attribute_tags:
    attributes.append(att.text)
attributes = attributes[1:]
print(len(attributes))

#Scraping year details
year_tags = driver.find_elements_by_xpath("//tbody/tr/td[7]/p")
for yr in year_tags:
    year.append(yr.text)
year = year[1:]
print(len(year))

uci_datasets = pd.DataFrame({'Dataset name': name,'Data Type':data_type,'Default Task':task,'Attribute Types':attribute_type,'No. of Instances':instances,'No. of Attributes':attributes, 'Year': year})
uci_datasets
```
